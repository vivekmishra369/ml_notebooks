```python
%matplotlib inline
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
rawdata = pd.read_csv("../datasets/smsspamcollection/SMSSpamCollection",
                      sep='\t',
                     names=["labels","message"])
rawdata.shape
```

```python
rawdata.head()
```

```python
corpus_stemmed = []
corpus_lemmatized = []
ps = PorterStemmer()
wordnet = WordNetLemmatizer()
for i in range(rawdata.shape[0]):
    sen = re.sub('[^a-zA-Z]',' ',rawdata['message'][i])
    sen = sen.lower()
    sen = sen.split()
    sen_stem = [ps.stem(word) for word in sen if word not in set(stopwords.words('english'))]
    sen_stem = ' '.join(sen)
    corpus_stemmed.append(sen_stem)
    sen_lemm = [wordnet.lemmatize(word) for word in sen if word not in set(stopwords.words('english'))]
    sen_lemm = ' '.join(sen_lemm)
    corpus_lemmatized.append(sen_lemm)
    
```
